```python
import pandas as pd
from datetime import datetime

# Example timestamp
timestamp = 1228137762.0

# Convert the Unix timestamp to a pandas datetime object
datetime_obj = pd.to_datetime(timestamp, unit='s')

# Print the datetime object
print("Pandas datetime:", datetime_obj)

# Convert to a Python datetime object
datetime_python = datetime_obj.to_pydatetime()

# Print the Python datetime object
print("Python datetime:", datetime_python)

# Format the datetime object if needed
formatted_datetime = datetime_python.strftime('%Y-%m-%d %H:%M:%S')

# Print the formatted datetime
print("Formatted datetime:", formatted_datetime)
```
